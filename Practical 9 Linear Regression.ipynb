# Develop an application to predict the best salary to be offered using Linear Regression

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#We are given dataset of list of salaries of employees at LMN Co Ltd the HR here has got a candidate 
#with 7 year of experience we need to predict the best salary that should be offered to him?
```

```python
#  data preprocessing
dataset = pd.read_csv("D:/Machine Learning/Employee_Salary_Data.csv")  # Importing the dataset.
dataset.head(10)
```

```python
x = dataset.iloc[:, :-1].values  #seperating dependent and independent columns
y = dataset.iloc[:, -1].values
```

```python
#data splitting
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
print(x_train, y_train)
```

```python
print (x_test , y_test)
```

```python
#regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Visualisation
plt.scatter(x_train, y_train, color = "red")
plt.grid()
```

```python
plt.scatter(x_train, regressor.predict(x_train), color="blue")  #Plotting graph between salary and experience in training data set.
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid()
```

```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid()
```

```python
plt.scatter(x_test, y_test, color = "red")  #Plotting graph between salary and experience of test data set.
plt.plot(x_train, regressor.predict(x_train), color="blue")
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid()
```

```python
#prediction
Y_pred1 = regressor.predict([[1]])
print(Y_pred1)
Y_pred2 = regressor.predict([[7]])
print(Y_pred2)
```

```python
y_pred3 = regressor.predict(x_test)
print(y_pred3)
```

```python
print (regressor.coef_)  #slope
print(regressor.intercept_)  #intercept
```

```python
print("R Squared: {:.2f}".format(regressor.score(x,y)*100))
```

```python
regressor.score(x_test, y_test)*100
```
